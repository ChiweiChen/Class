```python
def oddeven(answers):
    for answer in answers: 
        start= 0
        for i in range (1,answer+1):
            start=start+ i
        if start%2==1:
            print("odd")
        else:
            print("even")
        print(start)
answers=[3,10,13]
oddeven(answers)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import os
options= Options()
options.add_argument('--headless')
options.add_argument('--disable-gpu')
driver=webdriver.Chrome(os.getcwd()+"/chromedriver",chrome_options=options)

for i in range(1,3):
    driver.get('https://technews.tw/category/internet/page/1')
    time.sleep(0.5)
    sourceCode=BeautifulSoup(driver.page_source,"html.parser")
    contents=sourceCode.find_all("div","content")
    for content in contents:
        sourceCode2=BeautifulSoup(str(content),"html.parser")
        title=sourceCode2.find_all("h1","entry-title")[0].text
        print(title)
        date=sourceCode2.find_all("span","body")[1].text
        print(date)
        category=sourceCode2.find_all("span","body")[2]
        sourceCode3=BeautifulSoup(str(category),"html.parser")
        typs=sourceCode3.find_all("a")
        keywords=""
        for typ in typs:
            if typ==typs[-1]:
                keywords=keywords+typ.text
            else:
                keywords=keywords+typ.text+","
        print(keywords)
        links=sourceCode2.find_all("a",href=True)[0]['href']
        print(links)
        pretext=sourceCode2.find_all("p")[0]
        print(pretext)
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import os
options= Options()
options.add_argument('--headless')
options.add_argument('--disable-gpu')
driver=webdriver.Chrome(os.getcwd()+"/chromedriver",chrome_options=options)
f=open(os.getcwd()+"/data.txt",'w',encoding='utf8')
for i in range(1,3):
    pagenumber=str(i)
    driver.get('https://www.ptt.cc/bbs/LoL/index'+pagenumber+'.html')
    time.sleep(0.5)
    sourceCode=BeautifulSoup(driver.page_source,"html.parser")
    articles=sourceCode.find_all("div","r-ent")
    for article in articles:
        sourceCode2=BeautifulSoup(str(article),"html.parser")
        title=sourceCode2.find_all("div","title")[0].text
        title=title.strip()
        if title.startswith("[公告]"):
            pass
        else:
            date=sourceCode2.find_all("div","date")[0].text
            number=sourceCode2.find_all("div","nrec")[0].text
            author=sourceCode2.find_all("div","author")[0].text
            print(title.strip())
            print(author)
            print(date)
            number=number.replace("爆",">100")
            print(number)
            data=title+"\n"+number+"\n"+author+"\n"+date
            f.write(data)
f.close()
driver.quit()
        
```
